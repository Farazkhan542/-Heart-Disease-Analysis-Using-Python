```python
df=strat_train_set.drop("num",axis=1)
df_label=strat_train_set["num"].copy()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
my_pipeline=Pipeline([
    ("imputer",SimpleImputer(strategy="median")),
    ("std_scaler",StandardScaler()),
])
```

```python
df_num_tr=my_pipeline.fit_transform(df)
```

```python
df_num_tr
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model=DecisionTreeRegressor()
model=LinearRegression()
#model=RandomForestRegressor()
model.fit(df_num_tr,df_label)
```

```python
from sklearn.metrics import mean_squared_error
heart_disease_predictions=model.predict(df_num_tr)
lin_mse=mean_squared_error(df_label,heart_disease_predictions)
lin_rmse=np.sqrt(lin_mse)
```

```python
lin_mse
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,df_num_tr,df_label,scoring="neg_mean_squared_error",cv=10)
rmse_score=np.sqrt(-scores)
```

```python
rmse_score
```

```python
def print_scores(scores):
    print("Scores = ",scores)
    print("Mean = ",scores.mean())
    print("Standard deviation = ",scores.std())
```

```python
print_scores(rmse_score)
```

```python
from joblib import dump,load
dump(model,"heart_diseases.joblib")
```

```python
#testing my data
X_test=strat_test_set.drop("num",axis=1)
Y_test=strat_test_set["num"].copy()
X_test_prepared=my_pipeline.transform(X_test)
final_prediction=model.predict(X_test_prepared)
final_mse=mean_squared_error(Y_test,final_prediction)
final_rmse=np.sqrt(final_mse)
```
